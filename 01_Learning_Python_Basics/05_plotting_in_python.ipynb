```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
demo_data = pd.read_csv("data_in/day02_02/life_expectancy_and_income.csv")
```

```python
demo_data_2019 = demo_data.loc[demo_data["year"] == 2019].copy()
```

```python
demo_data.info()
```

```python
# plt.rcParams['figure.figsize'] = [5,3]
# plt.rcParams['figure.dpi'] = 120
# plt.rcParams['font.size'] = 10
```

```python
plt.scatter(x = "income_per_person", y = "life_expectancy", data = demo_data_2019)
plt.ylabel('Life expectancy') 
plt.title("Scatterplot of Income per person vs. life expectancy")
plt.xlabel('Income per person ($, ppp)') 
```
